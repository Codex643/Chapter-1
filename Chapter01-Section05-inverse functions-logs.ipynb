```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
# Define the domain for each function
x_exp = np.linspace(0.1, 3, 100)   
y_exp = np.exp(1- x_exp) 

x_inv = np.linspace(0.1, 3, 100) 
y_inv = 1 - np.log(x_inv)  

# Plotting
plt.plot(x_exp, y_exp, label="$f(x) = e[1-x]$", color='blue')
plt.plot(x_inv, y_inv, label="$f⁻¹(x) = 1 - ln(x)", color='green')
plt.plot(x_exp, x_exp, label="y = x (line of symmetry)", linestyle='--', color='gray')

plt.title("Exponential Function and Its Inverse")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.axis('equal')  # Keeps symmetry across y=x
plt.show()
```

```python
sym.init_printing
x, y = sym.symbols('x y')
f = (1+ sym.exp(-x))
eq= sym.Eq(y, f)
inverse_solved = sym.solve(eq, x)[0]
print ("equation:")
display(eq)
print ("inverse:")
display (inverse_solved)
```

```python
inverse_solved = sym.solve(eq, x)[0]
```

```python
f_lambdified = sym.lambdify(x, f, modules=['numpy'])
inv_lambdified = sym.lambdify(y, inverse_solved, 'numpy')
```

```python
x_vals = np.linspace(0.5, 20, 400)
y_vals = f_lambdified(x_vals)
inv_vals = inv_lambdified(y_vals)
```

```python
plt.plot(x_vals, y_vals, color='blue')
plt.plot(y_vals, inv_vals, color='red')
```
