- import `numpy` for numerical operations.

```python
import numpy as np
import matplotlib.pyplot as plt
```

- Get the set of x values from -a to -a with n values
- The range of x values are from -sqrt(5) to +sqrt(5), but there are numerical problems with this. (e.g. try a = 5**0.5)

```python
a = 2.236
npoints = 250
x = np.linspace(-a, a, npoints)
```

```python
#print(x) #Remove the first hashtag to see the values of x.
```

- Define the functionss

```python
ypos = (5-x**2)**0.5 + 3
```

```python
yneg = -(5-x**2)**0.5 + 3
```

```python
fig, ax = plt.subplots()
ax.set_aspect('equal', adjustable='box')
plt.plot(x, ypos, color='red')
plt.plot(x, yneg, color='green')
plt.title('Plot of $x^2 + (y-3)^2 = 5$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.show()
```
