```python
import numpy as np
import matplotlib.pyplot as plt
```

- Example of a reflection of f(x) = 1/x and f(x) = 1/x^2 about the x and y axes. Remember you have to manage the domain. 

```python
a = 1
npoints = 99
xpos = np.linspace(0.01, a, npoints)
xneg = -xpos  
```

```python
n = 1

y0 = (1/xpos)**(n)
y1 = (-1/xpos)**(n)
y2 = (1/xneg)**(n)
y3 = (-1/xneg)**(n) 
```

- Add a title, labels, grid, labels, and legend

```python
plt.plot(xpos, y0, label=r'$(1/xpos)^n, [0.1, 1.0]$', color='black')
plt.plot(xpos, y1, label=r'$(-1/xpos)^n, [0.1, 1.0]$', color='red')
plt.plot(xneg, y2, label=r'$(1/xneg)^n [-1.0, -0.1]$', color='green')
plt.plot(xneg, y3, label=r'$-(1/xneg)^n [-1.0, -0.1]$', color='blue')
plt.title('X,Y reflections')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
n = 2

y0 = (1/xpos)**(n)
y1 = (-1/xpos)**(n)
y2 = (1/xneg)**(n)
y3 = (-1/xneg)**(n) 
plt.plot(xpos, y0, label=r'$(1/xpos)^n, [0.1, 1.0]$', color='black')
plt.plot(xpos, y1, label=r'$(-1/xpos)^n, [0.1, 1.0]$', color='red')
plt.plot(xneg, y2, label=r'$(1/xneg)^n [-1.0, -0.1]$', color='green')
plt.plot(xneg, y3, label=r'$-(1/xneg)^n [-1.0, -0.1]$', color='blue')
plt.title('X,Y reflections')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
a = 1
npoints = 99
x = np.linspace(-a, a, npoints)  # Use a symmetric domain for exponentials

y0 = 3 ** x
y1 = 10 ** x
y2 = (1/3) ** x
y3 = (1/10) ** x

plt.figure(figsize=(8, 6))  # Set the figure size

plt.plot(x, y0, label=r'$3^x$', color='black')
plt.plot(x, y1, label=r'$10^x$', color='red')
plt.plot(x, y2, label=r'$(1/3)^x$', color='green')
plt.plot(x, y3, label=r'$(1/10)^x$', color='blue')
plt.title('Exponential Functions')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
a = 1
npoints = 99
x = np.linspace(-a, a, npoints)  # Use a symmetric domain for exponentials

y0 = np.e ** x - 2         # e^x shifted down by 2
y1 = np.exp(x - 2)         # e^(x-2), shifted right by 2
y2 = -np.exp(x)            # -e^x, reflection over x-axis
y3 = np.exp(-x)            # e^(-x), reflection over y-axis
y4 = -np.exp(-x)           # -e^(-x), reflection over both axes

plt.figure(figsize=(8, 6))  # Set the figure size

plt.plot(x, y0, label=r'$e^x - 2$', color='black')
plt.plot(x, y1, label=r'$e^{x-2}$', color='red')
plt.plot(x, y2, label=r'$-e^x$', color='green')
plt.plot(x, y3, label=r'$e^{-x}$', color='blue')
plt.plot(x, y4, label=r'$-e^{-x}$', color='purple')
plt.title('Exponential Function Transformations')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x)
def f(x):
    return (1 - np.exp(1/x)) / (1 + np.exp(1/x))

# Create x values excluding zero
x = np.linspace(-5, 5, 1000)
x = x[x != 0]  # Remove x = 0 to avoid division by zero

# Compute y values
y = f(x)

# Plotting
plt.style.use('seaborn-v0_8')
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$f(x) = \frac{1 - e^{1/x}}{1 + e^{1/x}}$', color='blue')
plt.axhline(0, color='gray', linestyle='--', linewidth=1)
plt.axvline(0, color='gray', linestyle='--', linewidth=1)
plt.title('Graph of f(x) = (1 - e^(1/x)) / (1 + e^(1/x))')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.tight_layout()
```
