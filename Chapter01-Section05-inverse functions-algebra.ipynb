```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x_vals = np.linspace(0, 1, 400)
y_vals = 1 - x_vals**2

y_vals_inv = np.linspace(0, 1, 400)
x_vals_inv = np.sqrt(1 - y_vals_inv)

plt.plot(x_vals, y_vals, label='f(x) = 1 - x²', color='red')
plt.plot(y_vals_inv, x_vals_inv, label='Inverse: f⁻¹(y)', color='green')
plt.plot(x_vals, x_vals, color='black', linestyle=':', label='y = x')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Function and Its Inverse')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Domain for g(x): x + 1 >= 0 => x >= -1
x_vals = np.linspace(-1, 8, 400)
y_vals = 2 + np.sqrt(x_vals + 1)

# Domain for inverse: y >= 2 (since sqrt is always >= 0)
y_inv = np.linspace(2, 5, 400)
x_inv = (y_inv - 2)**2 - 1

plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label=r'$g(x) = 2 + \sqrt{x+1}$', color='red')
plt.plot(x_inv, y_inv, label=r'$g^{-1}(y) = (y-2)^2 - 1$', color='green')
plt.plot(x_vals, x_vals, color='black', linestyle=':', label='$y = x$')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Function $g(x)$ and Its Inverse')
plt.legend()
plt.grid(True)
plt.axis('equal')
```
